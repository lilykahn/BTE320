```python
interest = float(input("Set interest value (%): "))
principal = float(input("Enter initial amount on the account ($): "))
n = int(input("Enter number of years: "))

final = principal * (1 + interest/100) ** n

print(f"Total amount after {n} years: {final}$")
```
