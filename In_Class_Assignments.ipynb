```python
numb1 = float(input("Enter the first number: "))
numb2 = float(input("Enter the second number: "))
operator = input("Enter the operator (+, -, *, /): ")

if operator == "+":
  result = numb1 + numb2
  print("Result:", result)

elif operator == "-":
  result = numb1 - numb2
  print("Result:", result)

elif operator == "(*)":
  result = numb1 * numb2
  print("Result:", result)

elif operator == "/":
  if numb2 != 0:
    result = numb1 / numb2
    print("Result:", result)
  else:
    print("Error: cannot devide by zero")

else:
  print("Error: invalid operator")
```

```python
import random

play = input("Do you want to play a game (y/n)? ")

while play == 'y':
  player = input("Choose b/w rock, paper, scissors: ")
  computer = random.choice(["rock", "paper", "scissors"])

  if player == computer:
    print("Draw")
  elif player == "rock" and computer == "scissors":
    print("Player wins!")
  elif player == "scissors" and computer == "paper":
    print("Player wins!")
  elif player == "paper" and computer == "rock":
    print("Player wins!")
  else:
    print("Computer wins!")

  play = input("Do you want to continue (y/n)? ")
```

```python
numbers = []

while True:
  user_input = float(input("Enter a number (enter 0 to stop): "))
  if user_input == 0:
    break

  numbers.append(user_input)

numbers.sort(reverse=True)

print("Numbers in descending order:")
print(numbers)
```
