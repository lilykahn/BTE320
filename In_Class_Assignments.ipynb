```python
def checkEmailAddress(email):
    parts = email.split("@")

    if len(parts) != 2:
        return False

    username = parts[0]
    domain_parts = parts[1].split(".")

    if username == "":
        return False

    if len(domain_parts) < 2:
        return False

    domain = domain_parts[0]
    topdomain = domain_parts[-1]

    if domain == "" or topdomain == "":
        return False

    return True

email = input("Enter email address: ")
print(checkEmailAddress(email))
```

```python
def getValues(n):
    items = []
    for i in range(n):
        while True:
            try:
                value = float(input("Enter a value: "))
                items.append(value)
                break
            except ValueError:
                print("Invalid input. Please enter a number.")
    return items


def calcAverage(alist):
    assert isinstance(alist, list), "'alist' must be a list"

    if len(alist) < 3:
        raise ValueError("The list must contain at least 3 items")

    trimmed_list = alist[3:]

    if len(trimmed_list) == 0:
        raise ValueError("Not enough values after excluding first three")

    return sum(trimmed_list) / len(trimmed_list)


def main():
    try:
        n = int(input("Enter the number of items: "))

        if n < 3:
            print("You need at least 3 values.")
            return

        alist = getValues(n)
        avg = calcAverage(alist)

        print("Average (excluding first 3 values):", avg)

    except ValueError as e:
        print("Error:", e)

main()
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        dx = (self.x - other.x) ** 2
        dy = (self.y - other.y) ** 2
        return (dx + dy) ** 0.5


class Rectangle:
    def __init__(self, p1, p2, p3, p4):
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3
        self.p4 = p4

    def width(self):
        return self.p1.distance(self.p2)

    def height(self):
        return self.p1.distance(self.p3)

    def area(self):
        return self.width() * self.height()
```
