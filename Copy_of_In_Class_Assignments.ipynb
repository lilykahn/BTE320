```python
numb1 = float(input("Enter the first number: "))
numb2 = float(input("Enter the second number: "))
operator = input("Enter the operator (+, -, *, /): ")

if operator == "+":
  result = numb1 + numb2
  print("Result:", result)

elif operator == "-":
  result = numb1 - numb2
  print("Result:", result)

elif operator == "(*)":
  result = numb1 * numb2
  print("Result:", result)

elif operator == "/":
  if numb2 != 0:
    result = numb1 / numb2
    print("Result:", result)
  else:
    print("Error: cannot devide by zero")

else:
  print("Error: invalid operator")
```
