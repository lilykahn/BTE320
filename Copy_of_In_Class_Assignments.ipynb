```python

sales = float(input("Enter sales value ($): "))

COGS = sales * 0.4

GP = sales - COGS

tax = float(input("Enter tax rate (%): "))

taxes = GP * tax/100

Net_Income = GP - taxes

print(f"Sales: {sales}$")
print(f"Cost of Goods Sold: {COGS}$ ")
print(f"Gross Profit: {GP}$")
print(f"Taxes: {taxes}$")
print(f"Net_Income: {Net_Income}$")
```
